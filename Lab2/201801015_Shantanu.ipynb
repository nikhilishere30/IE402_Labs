```python
# Import the libraries
import pulp as p 
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
from sympy.solvers import solve
```

```python
   def LPP_Solver():
    
    # Extra Variables for later calculations
    constraint = []
    d = np.linspace(0,25,1000)
    #dd = np.linspace(-10,50,1000)
    xx,yy = np.meshgrid(d,d)
    yLine = []
    X = sym.Symbol('x')
    Y = sym.Symbol('y')
    yFunc = []
    xFunc = []
    pts = [(-0.6969,-0.6969) for i in range(10)]
    cc = [(0,0,0,'') for i in range(3)]
    
    # Problem Variables  
    x = p.LpVariable("x", lowBound = 0)   # Create a variable x >= 0 
    y = p.LpVariable("y", lowBound = 0)   # Create a variable y >= 0 
    
    # Inputs
    case = input('Maximize/Minimize?: ').lower()
    objC1, objC2 = input('X and Y coefficients of Objective function seperated by spaces: ').split(' ')
    objective = int(objC1)*x + int(objC2)*y
    o1 = int(objC1)
    o2 = int(objC2)
    constraints = []
    for i in range(3):
        conC1, conC2, val, sign = input('X and Y coefficients and sign(geq/leq) of Constraint '+ str(i+1) + ' seperated by spaces: ').split(' ')
        cc[i] = [int(conC1), int(conC2), int(val), sign]
        if sign == 'geq':
            constraints.append(int(conC1)*x + int(conC2)*y >= int(val))
            constraint.append(int(conC1)*xx + int(conC2)*yy >= int(val))
        elif sign == 'leq':
            constraints.append(int(conC1)*x + int(conC2)*y <= int(val))
            constraint.append(int(conC1)*xx + int(conC2)*yy <= int(val))
        yLine.append((int(val)-int(conC1)*d)/int(conC2))
        yFunc.append((int(val)-int(conC1)*X)/int(conC2))
        xFunc.append((int(val)-int(conC2)*Y)/int(conC1))
        pts[i+1] = [0,yFunc[i].evalf(subs={X: 0})]
        pts[i+4] = [xFunc[i].evalf(subs={Y: 0}),0]
    
    # Create a LP Minimization problem 
    if case=='mazimize':
        Lp_prob = p.LpProblem('Problem', p.LpMaximize)  
    elif case=='minimize':
        Lp_prob = p.LpProblem('Problem', p.LpMinimize) 

    # Objective Function 
    Lp_prob += objective 

    # Constraints
    for c in constraints:
        Lp_prob += c

    # Display the problem 
    # print(Lp_prob) 
    
    # Display the output
    print('\033[1m' + 'OUTPUT' + '\033[0m')
    
    # LPP Solver
    status = Lp_prob.solve() 
    
    # Finding Extreme points at max 10 possible, so we calculate these 10 points and then check if they satisfy constraints
    pts[0] = [0,0]
    #now intersect the three lines with each other
    #check for parallel lines
    #check for lines perpendicular to axes
    x1 = solve(yFunc[0]-yFunc[1])[0]
    x2 = solve(yFunc[1]-yFunc[2])[0]
    x3 = solve(yFunc[2]-yFunc[1])[0]
    pts[7] = [x1,yFunc[0].evalf(subs={X: float(x1)})]
    pts[8] = [x2,yFunc[1].evalf(subs={X: float(x2)})]
    pts[9] = [x3,yFunc[2].evalf(subs={X: float(x3)})]
    
    i=len(pts)-1
    while(i>=0):
        for c in cc:
            if c[3] == 'geq':
                if c[0]*pts[i][0] + c[1]*pts[i][1] < c[2]:
                    pts.remove(pts[i])
                    #i=i-1
                    #x=x-1
                    continue
            elif c[3] == 'leq':
                if c[0]*pts[i][0] + c[1]*pts[i][1] > c[2]:
                    pts.remove(pts[i])
                    #i=i-1
                    #x=x-1
                    continue
        i-=1
    #remove duplicates
    final = []
    i=len(pts)-1
    while(i>=0):
        if pts[i] not in final: 
            final.append(pts[i])
        i-=1
    pts = final
    #sorted output
    pts=sorted(pts, key = lambda x: x[0], reverse=True)
    
    # Printing the final solution 
    if status == -1:
        print('Infeasible Region')
        print('No Solution')
    elif status == -2:
        print('Unbounded Feasible Region')
        for pt in pts:
            print('Extreme point: ('+str(round(pt[0],2))+','+ str(round(pt[1],2)) +') Value: '+ str(round(o1*pt[0]+o2*pt[1],2)))
        if case=='mazimize':
            print('Multiple maximum solution')
            print('Maximum Solution: ' + 'x = ' + str(p.value(x)) + ', y = ' + str(p.value(y)) + ', value = ' + str(p.value(Lp_prob.objective)))
        elif case=='minimize':
            print('Unique minimum solution')
            print('Minimum Solution: ' + 'x = ' + str(p.value(x)) + ', y = ' + str(p.value(y)) + ', value = ' + str(p.value(Lp_prob.objective)))
        
    elif status == 1:
        print('Bounded Feasible Region')
        for pt in pts:
            print('Extreme point: ('+str(round(pt[0],2))+','+ str(round(pt[1],2)) +') Value: '+ str(round(o1*pt[0]+o2*pt[1],2)))
        if case=='mazimize':
            print('Unique maximum solution')
            print('Maximum Solution: ' + 'x = ' + str(p.value(x)) + ', y = ' + str(p.value(y)) + ', value = ' + str(p.value(Lp_prob.objective)))
        elif case=='minimize':
            print('Unique minimum solution')
            print('Minimum Solution: ' + 'x = ' + str(p.value(x)) + ', y = ' + str(p.value(y)) + ', value = ' + str(p.value(Lp_prob.objective)))
        
    else:
        print(p.LpStatus[status])
        
    # Plot
    plt.figure(figsize=(10,10));
    plt.grid(True, which='both')
    
    # extreme points
    for pt in pts:
        plt.plot(pt[0],pt[1],'go',markersize=8)
    
    # Lines
    plt.plot(d,yLine[0],'k--')
    plt.plot(d,yLine[1],'k--')
    plt.plot(d,yLine[2],'k--')
    
    # Feasible Region
    plt.imshow( (constraint[0] & constraint[1] & constraint[2]).astype(int), extent=(xx.min(),xx.max(),yy.min(),yy.max()),origin="lower", cmap="Greys", alpha = 0.3);
```
